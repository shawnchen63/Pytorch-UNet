```python
IMG_EXTENSIONS = [
    '.jpg', '.JPG', '.jpeg', '.JPEG',
    '.png', '.PNG', '.ppm', '.PPM', '.bmp', '.BMP',
]

def is_image_file(filename):
    return any(filename.endswith(extension) for extension in IMG_EXTENSIONS)
```

```python
import os
input_paths = []
output_paths = []
target_path = '/home/shchen/Pytorch-UNet/data/test_real_data'
for root, _, fnames in os.walk('/home/shchen/DA_dahazing/datasets/dehazing/unlabeled'):
    for fname in fnames:
        if is_image_file(fname):
            path = os.path.join(root, fname)
            input_paths.append(f'{path}')
            output_paths.append(os.path.join(target_path, fname))
```

```python
input_paths = " ".join(sorted(input_paths)[:100])
output_paths = " ".join(sorted(output_paths)[:100])
```
