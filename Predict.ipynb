```python
IMG_EXTENSIONS = [
    '.jpg', '.JPG', '.jpeg', '.JPEG',
    '.png', '.PNG', '.ppm', '.PPM', '.bmp', '.BMP',
]

def is_image_file(filename):
    return any(filename.endswith(extension) for extension in IMG_EXTENSIONS)
```

```python
import os
input_paths = []
output_paths = []
#target_path = '/home/shchen/Pytorch-UNet/test_results/LOL_pad/'
input_path = '/home/shchen/Pytorch-UNet/data/test_data/test_data-without_GT/'
#input_path = '/mnt/sdb/sihao/Resutls_SOTA_methods/testing_dataset/SCIE_testing_dataset/part2-resize/low_resize'
#input_path =  '/home/shchen/Pytorch-UNet/data/test_data/SCIE'
#input_path = '/mnt/sdb/sihao/our485/low'
for root, _, fnames in os.walk(input_path):
    for fname in fnames:
        if is_image_file(fname):
            path = os.path.join(root, fname)
            #fname_index = fname.split(".")[0]
            input_paths.append(f'{path}')
            output_paths.append(path.replace("test_data-without_GT", "test_data-without_GT_noise_pad"))
            #output_paths.append(target_path+fname)
```

```python
from PIL import Image
from eval import *
import torch
import numpy as np

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def evaluate(img_path, gt_path):
    #print(img_path)
    #print(gt_path)
    img = Image.open(img_path).convert('RGB')
    gt = Image.open(gt_path).convert('RGB')
#    img = img.resize((gt.size[0], gt.size[1]))
#    gt = gt.resize((512,512))
    
    img = np.array(img).astype(np.float32) / 255
    gt = np.array(gt).astype(np.float32) / 255
        
    img = np.moveaxis(img, -1, 0)
    gt = np.moveaxis(gt, -1, 0)
    
    img = np.expand_dims(img, axis=0)
    gt = np.expand_dims(gt, axis=0)

    img = torch.from_numpy(img)
    gt = torch.from_numpy(gt)
    
    print(img.size())
    
    MSE = nn.MSELoss()
    return {"SSIM":SSIM(img,gt).item(), "PSNR":PSNR(img,gt), "MSE":MSE(img,gt).item()}    
```

```python
def evaluate_results(output_paths):
    ssim = 0
    psnr = 0
    mse = 0
    for i in range(len(output_paths)):
        temp = evaluate(output_paths[i], input_paths[i].replace("low","label"))
        #print(temp)
        ssim += temp["SSIM"]/len(output_paths)
        psnr += temp["PSNR"]/len(output_paths)
        mse += temp["MSE"]/len(output_paths)
    return ssim,psnr,mse
```
